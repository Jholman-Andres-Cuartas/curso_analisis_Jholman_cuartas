## Displot

```python
df = sns.load_dataset("mpg")
df.head()
```

```python
df.sort_values('weight', ascending=False)
```

```python
sns.displot(df["weight"])
```

```python
sns.displot(df["weight"], kde=True)
```

```python
sns.displot(df["weight"], bins=5)
```

```python
df = sns.load_dataset("tips")
sns.displot(df, x="day")
```

```python
sns.displot(df, x="day", shrink=.8)
```

```python
sns.displot(df, x="size")
```

```python
sns.displot(df, x="size", discrete=True)
```

```python
df = sns.load_dataset("penguins")
sns.displot(df, x="flipper_length_mm")
```

```python
sns.displot(df, x="flipper_length_mm", hue="species")
```

```python
sns.displot(df, x="flipper_length_mm", col="sex")
```

```python
sns.displot(df, x="flipper_length_mm", kind="kde")
```

```python
sns.displot(df, x="flipper_length_mm", hue="species", kind="kde")
```

```python
sns.displot(df, x="bill_length_mm", y="bill_depth_mm")
```

```python
sns.displot(df, x="bill_length_mm", y="bill_depth_mm", hue="species")
```

## Jointplot

```python
sns.relplot(data=df, x="total_bill", y="tip", hue="smoker", col="time")
```

```python
sns.relplot(data=df, x="total_bill", y="tip", hue="smoker", col="time", row="sex")
```

```python
sns.relplot(data=df, x="total_bill", y="tip", hue="smoker", col="time", size="size")
```

## Rugplot

```python
df = sns.load_dataset("tips")
sns.rugplot(df["total_bill"])
```

```python
sns.kdeplot(data=df, x="total_bill")
sns.rugplot(data=df, x="total_bill")
```

```python
sns.scatterplot(data=df, x="total_bill", y="tip")
sns.rugplot(data=df, x="total_bill", y="tip")
```

```python
sns.scatterplot(data=df, x="total_bill", y="tip")
sns.rugplot(data=df, x="total_bill", y="tip", height=.1)
```

## Barplot

```python
df = sns.load_dataset("tips")
sns.barplot(x="sex", y="total_bill", data=df)
```

```python
sns.barplot(x="sex", y="total_bill", data=df, estimator=np.cov) # mean() by default
```

## Countplot

```python
df = sns.load_dataset("tips")
sns.countplot(x="sex", data=df)
```

```python
ax = sns.countplot(x='sex', data=df)
ax.bar_label(ax.containers[0])
```

## Boxplot

```python
df = sns.load_dataset("tips")
sns.boxplot(x="day", y="total_bill", data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df, hue="sex")
plt.legend(loc=2)
```

## Violinplot

## Stripplot

## Swarmplot

```python
df = sns.load_dataset("tips")
sns.swarmplot(x="day", y="total_bill", data=df)
```

```python
sns.swarmplot(x="day", y="total_bill", data=df, hue='sex')
```

```python
sns.swarmplot(x="day", y="total_bill", data=df, hue='sex', palette='viridis')
```

```python
sns.violinplot(x="day", y="total_bill", data=df)
sns.swarmplot(x="day", y="total_bill", data=df, color="white")
```

## Heatmap

```python
df = sns.load_dataset("tips")
df_corr = df.corr(numeric_only=True)
sns.heatmap(df_corr, annot=True, cmap="viridis")
```

```python
df = sns.load_dataset("penguins")
df_corr = df.corr(numeric_only=True)
sns.heatmap(df_corr, annot=True, cmap="viridis")
```

```python
sns.heatmap(df_corr, annot=True, linewidth=.5, cmap='viridis')
```

```python
df = sns.load_dataset("flights")
df.head()
```

```python
df_pivot = df.pivot_table(index="month", columns="year", values="passengers")
df_pivot.head()
```

```python
sns.heatmap(df_pivot, cmap="viridis", linecolor="white", linewidth=1)
```
